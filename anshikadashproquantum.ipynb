The combined and processed data is now in `final_df`. This DataFrame contains 'Sales Date', 'Sales', and 'Region' for all 'Pink Morsel' transactions from the three input files. We can now proceed with further analysis or visualization if needed.

```python
import pandas as pd

# Define the file paths for the sales data
file_paths = [
    '/daily_sales_data_0.csv',
    '/daily_sales_data_1.csv',
    '/daily_sales_data_2.csv'
]

# List to store processed dataframes
processed_dfs = []

# Loop through each file, process it, and append to the list
for file_path in file_paths:
    # Load the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)

    # Clean 'price' column by removing '$' and converting to numeric
    # Using .str.replace() to ensure string replacement of literal '$'
    df['price'] = df['price'].str.replace('$', '', regex=False).astype(float)

    # Filter for 'Pink Morsel' products (case-insensitive)
    df_filtered = df[df['product'].str.lower() == 'pink morsel'].copy()

    # Calculate the 'sales' column
    df_filtered['sales'] = df_filtered['quantity'] * df_filtered['price']

    # Select and rename the required columns
    df_processed = df_filtered[['date', 'sales', 'region']].copy()
    df_processed.rename(columns={'date': 'Sales Date', 'sales': 'Sales', 'region': 'Region'}, inplace=True)

    # Append the processed DataFrame to the list
    processed_dfs.append(df_processed)

# Concatenate all processed DataFrames into a single one
final_df = pd.concat(processed_dfs, ignore_index=True)

# Display the first 5 rows of the final DataFrame
display(final_df.head())
```

```python
final_df.to_csv('final_df.csv', index=False)
print("DataFrame saved to final_df.csv")

# Provide a link for download (this will only work in a Colab environment)
from google.colab import files
files.download('final_df.csv')
```
